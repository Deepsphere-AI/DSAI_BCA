Complex Numbers : A complex number is the sum of a real number and an imaginary number. A complex number is of the form a + ib and is usually represented by z. Here both a and b are real numbers. The value 'a' is called the real part which is denoted by Re(z), and 'b' is called the imaginary part Im(z).  Also, ib is called an imaginary number.

image.png

```python
vAR_Complex_No = 5 +8j
vAR_Complex_No = complex(5,8)
print('The Real Part the Complex Number is = ', vAR_Complex_No.real)
print('The Real Part the Complex Number is = ', vAR_Complex_No.imag)
```

```python
vAR_c1 = 4 + 5j
vAR_c2 = 3 + 5j

#Addition of Complex Numbers
print("Addition of two complex number =", vAR_c1 + vAR_c2)

#Subtraction of Complex Numbers
print("Subtraction of two complex number =", vAR_c1 - vAR_c2)

#Multiplication of Complex Numbers
print("Multiplication of two complex number =", vAR_c1 * vAR_c2)

#Division of Complex Numbers
print("Division of two complex number =", vAR_c1 / vAR_c2)
```

The phase of a complex number is the angle between the real axis and the vector representing the imaginary part

```python
import cmath, math, numpy
vAR_c = 4+ 4j
# phase
vAR_phase = cmath.phase(vAR_c)
print('4+ 4j Phase =', vAR_phase)
print('Phase in Degrees =', numpy.degrees(vAR_phase))

vAR_c = -4-4j
# phase
vAR_phase = cmath.phase(vAR_c)
print('-4- 4j Phase =', vAR_phase)
print('Phase in Degrees =', numpy.degrees(vAR_phase))
```

Booleans represent one of two values: True or False. You can evaluate any expression in Python, and get one of two answers, True or False.

When you compare two values, the expression is evaluated and Python returns the Boolean answer

```python
print("The answer:",100 > 90)
print("The answer:",100 == 90)
print("The answer:", 100 < 90)
```

When you run a condition in an if statement, Python returns True or False:

```python
vAR_A = 200
vAR_B = 33

if vAR_B > vAR_A:
  print("vAR_B is greater than vAR_B")
else:
  print("vAR_B is not greater than vAR_B")
  
```
