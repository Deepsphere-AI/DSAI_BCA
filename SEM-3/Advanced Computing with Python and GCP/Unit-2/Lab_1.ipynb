#***In Google Colab, develop a Simple Python Program for the conditional statements***

Conditional statements help you to make a decision based on certain conditions. These conditions are specified by a set of conditional statements having boolean expressions which are evaluated to a boolean value of true or false.

1. The If Statement : The if statement is a conditional statement in python, that is used to determine whether a block of code will be executed or not. Meaning if the program finds the condition defined in the if statement to be true, it will go ahead and execute the code block inside the if statement

image.png

```python
vAR_NOS = [2,4,6,9,5]

for vAR_i in vAR_NOS:
 
        # check if the given number is odd
        if vAR_i%2!=0:
          print(vAR_i,"is an odd number") # if condition is True print "odd"
 
        # check if no. is even
        if vAR_i%2==0:
          print(vAR_i,"is an even number") # if condition is false print "even"
          
```

2. If Else Statement : As discudded in the above section, The if statement executes the code block when the condition is true. Similarly, the else statement works in conjuncture with the if statement to execute a code block when the defined if condition is false

image.png

```python
vAR_NOS = [2,4,6,9,5]

for vAR_i in vAR_NOS:
 
        # check if the given number is odd
        if vAR_i%2!=0:
          print(vAR_i,"is an odd number") # if condition is True print "Odd"

        else:
          print(vAR_i,"is an even number") # if condition is True print "Even"
```

3. if-elif-else ladder: The elif statement is used to check for multiple conditions and execute the code block within if any of the conditions evaluate to be true.

The elif statement is similar to the else statement in the context that it is optional but unlike the else statement, there can be multiple elif statements in a code block following an if statement.

image.png

```python
# take a sample string

vAR_string ="Python"
 
# check if value in the string variable matches "Pyton"
if vAR_string == "Pyton":
    print ("The first condition is true")
 
# check if value in the string variable matches "Pthon"
elif vAR_string == "Pthon":
    print("The second condition is true")
 
# check if value in the string variable matches "Pytho"
elif vAR_string == "Pytho":
    print("The third condition is true")
 
# check if value in the string variable matches "Python"
elif vAR_string=="Python":
    print("The fourth condition is true")

else:
    print ("All the above conditions are false")
```

4. Nested if Statements:
A nested if statement is considered as if within another if statement(s).  These are generally used to check for multiple conditions.

image.png

```python
vAR_NOS = [4,5,9,17,21]
 
# for loop to iterate through
# the list and check each
# element of the list
for vAR_i in vAR_NOS:
 
    # condition1: check if no. is odd if yes execute the code block inside the first if statement
    if vAR_i%2!=0:
 
        # condition2: check if no. is also divisible by 3.
        if vAR_i%3==0:
 
            # if condition2 is true execute the below code block
            print (vAR_i,"is an odd number & divisible by 3")
 
        # if condition2 if False execute the below code block
 
        else:
            print (vAR_i, "is an odd number but not divisible by 3")
 
    # if condition1 is False execute the below code block
    else:
        print(vAR_i,"is an even number")
```
