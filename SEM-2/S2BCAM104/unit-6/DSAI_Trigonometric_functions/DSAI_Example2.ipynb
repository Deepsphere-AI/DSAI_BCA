ASIN()

    Returns the arc sine of the passed value X. The result will be a float value from -pi/2 to pi/2. 

```python
df['sin']  = ''
df['asin'] = ''

# math.asin() needs the passed value x to be inbetween -1 and 1.
for i in range(0,len(df)):
  df.sin[i] = math.sin(math.radians(df.elevation_angle[i]))
  df.asin[i] = math.asin(df.sin[i])

df
```
