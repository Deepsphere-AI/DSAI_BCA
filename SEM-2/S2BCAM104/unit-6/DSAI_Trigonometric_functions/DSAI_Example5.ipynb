ATAN()

```python
df['Tan_angle'] = ''
df['Wall_height'] = ''

# The following lines compute the height of each wall provided the elevation height and angle
for i in range(0,len(df)):
  #math.radians
  #math.tan
  df.Wall_height[i] = math.fabs((df.view_height[i])*(math.tan(math.pi/math.radians(df.elevation_angle[i]))))
  df.Tan_angle[i] = math.fabs(math.tan(math.pi/math.radians(df.elevation_angle[i])))
df
```
