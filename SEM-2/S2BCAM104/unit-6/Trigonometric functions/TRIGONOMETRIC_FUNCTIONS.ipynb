```python
import pandas as pd
import math
data = {'view_height':[90,90,90],
        'Elevation_angle':[30,60,36]}
df = pd.DataFrame(data)
df
```

```python
df['Tan_angle'] = ''
df['Wall_height'] = ''
for i in range(0,len(df)):
  #math.radians
  #math.tan
  df.Wall_height[i] = math.fabs((df.view_height[i])*(math.tan(math.pi/math.radians(df.Elevation_angle[i]))))
  # df.Wall_height[i] = math.fabs(df.Wall_height[i])
  df.Tan_angle[i] = math.fabs(math.tan(math.pi/math.radians(df.Elevation_angle[i])))
  # df.Tan_angle[i] = math.fabs(df.Tan_angle[i])
```
