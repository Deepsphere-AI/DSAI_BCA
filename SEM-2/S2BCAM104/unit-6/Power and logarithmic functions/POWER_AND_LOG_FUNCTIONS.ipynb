POWER AND LOG FUNCTIONS

```python
import pandas as pd
import math
data = {'Cars':['KIA','Duster','Ford fiesta'],
        'Ratings':[10,9,6]}
df = pd.DataFrame(data)
df
```

```python
df['exp'] = ''
df['expm'] = ''
df['log'] = ''
df['power'] = ''
df['sq_rt'] = ''
for i in range(0,len(df)):
  df.exp[i] = math.exp(df.Ratings[i])
  df.expm[i] = math.expm1(df.Ratings[i])
  df.log[i] = math.log(df.Ratings[i])
  df.power[i] = math.pow(df.Ratings[i],3)
  df.sq_rt[i] = math.sqrt(df.Ratings[i])
```
