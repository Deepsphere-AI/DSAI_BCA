Example 2 : WaveForms

```python
from scipy.signal import chirp, spectrogram
import matplotlib.pyplot as plt
t = np.linspace(4, 8, 300)
w = chirp(t, f0=4, f1=2, t1=5, method='linear')
plt.plot(t, w)
plt.title("Linear Chirp")
plt.xlabel('time in sec)')
plt.show()
```

Example 3: Guassian Filtering for Blurring Images

```python
import numpy as np
from scipy import signal, misc
import matplotlib.pyplot as plt

image = misc.ascent()
w = signal.windows.gaussian(51, 10.0)
#Apply convolution
image_new = signal.sepfir2d(image, w, w)


image = misc.ascent()
w = signal.windows.gaussian(51, 10.0)
image_new = signal.sepfir2d(image, w, w)

#Original Image
plt.figure()
plt.imshow(image)
plt.gray()
plt.title('Original image')
plt.show()

#Filtered Image
plt.figure()
plt.imshow(image_new)
plt.gray()
plt.title('Filtered image')
plt.show()
```

Example 4 :Removal of constant and linear trend of a second-order polynomial time series


```python
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt

t = np.linspace(-10, 10, 20)
y = 1 + t + 0.01*t**2
yconst = signal.detrend(y, type='constant')
ylin = signal.detrend(y, type='linear')

plt.plot(t, y, '-rx')
plt.plot(t, yconst, '-bo')
plt.plot(t, ylin, '-k+')
plt.grid()
plt.legend(['signal', 'const. detrend', 'linear detrend'])
plt.show()
```
